```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/iris-data.csv')
```

```python
print(df.head())
```

```python
x = df ['sepal length'].values
```

```python
y = df['sepal width'].values
```

```python
x = np.array(x)
y = np.array(y)
```

```python
n = len(x)
sum_x = np.sum(x)
sum_y = np.sum(y)
sum_xy = np.sum(x * y)
sum_xx = np.sum(x * x)

slope = (n * sum_xy - sum_x * sum_y) / (n * sum_xx - sum_x * sum_x)
intercept = (sum_y - slope * sum_x) / n
```

```python
#y_pred(sepal-width)
#x(sepal-length__independent)
y_pred = slope * x + intercept
```

```python
mean_y_pred = np.mean(y_pred)
```

```python
plt.scatter(x, y, color = "red")
plt.plot(x, y_pred, color = "green")
plt.xlabel("sepal-length")
plt.ylabel("sepal-width")
plt.title("Linear Regression")
plt.show()
```

```python
sum_nu = np.sum((y - y_pred) ** 2)
sum_de = np.sum((y - mean_y_pred) ** 2)
r_squared = 1- (sum_nu / sum_de)
print(r_squared)
```

```python
# prompt: calculate mean square ,mean absolute and root square

import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(y, y_pred)

# Calculate Mean Absolute Error (MAE)
mae = mean_absolute_error(y, y_pred)

# Calculate Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)

print("Mean Squared Error (MSE):", mse)
print("Mean Absolute Error (MAE):", mae)
print("Root Mean Squared Error (RMSE):", rmse)
```
